````python
# 13. Пользователь поочередно вводит 2 строки. Определите, какая строка длиннее и на сколько символов, а какая строка стоит раньше в лексикографическом порядке.
# Собрать результирующую строку, которая бы содержала 1 и 2 строки, разделенные переносом строки.

str1 = input(str("Введите стррку 1: "))
str2 = input(str("Введите строку 2: "))
len1 = len(str1)
print(f"Длина 1 строки: {len1}")
len2 = len(str2)
result = ' '
print(f"Длина 2 строки: {len2}")
if str1<str2: # проверка на лексикографический порядок
    result = "Вторая строка меньше в лексикографическом порядке"
elif str2 == str1:
    result = "Первая строка равна второй в лексикографическом порядке"
else:
    result = "Первая строка меньше в лексикографическом порядке"

# проверка разницы длин между строками
print(f"Первая строка длиннее на {abs(len1 - len2)} символов.\n{result}")

# строка, которая содержит 2 исходные вместе с пробелом
print(f"Результирующая строка:  {str1 + ' ' + str2}")
````

````python
# 14. Реализовать калькулятор, который принимает от пользователя через пробел строку следующего вида: «a op b»,
# где a и b – некоторые числа, а ‘op’ определяет оператор и может принимать значения «+,-,*,/,**,%». 
# В зависимости от оператора с помощью форматирования строк вывести результат в виде: «a + b = 3 + 2 = 5», где была получена строка «3 + 2». 
#Для операций возведения в степень и деления по модулю использовать вместо знака оператора соответствующие выражения. 
# Например, для строки «2 ** 3» должно быть выведено «a в степени b = 2 в степени 3 = 8».

input_str = input("Введите выражение (a op b): ")

# разбиение строку на операнды и оператор
tokens = input_str.split()
if len(tokens) != 3:
    print("Некорректный формат ввода.")
else:
    a = float(tokens[0])
    op = tokens[1]
    b = float(tokens[2])

    # вычисляется результат в зависимости от оператора
    if op == '+':
        result = a + b
    elif op == '-':
        result = a - b
    elif op == '*':
        result = a * b
    elif op == '/':
        if b != 0:
            result = a / b
        else:
            print("Деление на ноль недопустимо.")
            result = None
    elif op == '**':
        result = a ** b
    elif op == '%':
        result = a % b
    else:
        print("Неподдерживаемый оператор.")
        result = None

    # результат с форматированием
    if result is not None:
        print(f"{a} {op} {b} = {input_str} = {result}")
````

````python
# 18. Даны m (m>1) слов. Найти общее количество заданной буквы в этих словах.

words = input("Введите слова через запятую: ").split(',')

letter = input("Введите букву для поиска: ")

count = 0

for word in words:
    count += word.strip().lower().count(letter.lower())

print(f"Общее количество букв '{letter}' в словах: {count}")
````

````python
# 19. Даны m (m>1) слов. В каком из них доля (в %) заданной буквы больше.

words = input("Введите слова через запятую: ").split(',')
letter = input("Введите букву для анализа: ")

max_word = ""
max_percentage = 0

for word in words:
    word_lower = word.strip().lower()
    count_letter = word_lower.count(letter.lower())
    total_characters = len(word_lower)
    
    # процент доли буквы в слове
    percentage = (count_letter / total_characters) * 100
    
    # больше ли максимальной
    if percentage > max_percentage:
        max_word = word
        max_percentage = percentage

print(f"Слово с наибольшей долей буквы '{letter}' ({max_percentage:.2f}%) это: {max_word}")
````
